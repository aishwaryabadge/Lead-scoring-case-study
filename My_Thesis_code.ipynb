## ***Data Description***



**The dataset cointains the following features:**

The data set provided had features which are more or less related to the purchases of the users. All features are mentioned below with explanations.

The data set provided for model making has a total entry of "12330" with "18" features. Among these features 9 features are numerical, continuous and distinct, and 9 are categorical including the target feature Revenue.

The categorical features are Special Day, Month, Operating Systems, Browser, Region, Traffic Type, Visitor Type, Weekend. The numerical features are Administrative, Administrative Duration, Informational, Information Duration, Product Related, Product Relation duration, Bounce rates, Exit rates, Page Values.

**1.Administrative:** Number of pages visited by the user for user account management related activities.

**2.Administrative Duration:** Time spent on Admin pages by the user.

**3.Informational:** Number of pages visited by the user about the website

**4.Informational Duration:** Time spent on Informational pages by the user

**5.Product Related:** Number of product related pages visited by the user

**6.Product Related Duration:** Time spent on Product related pages by the user.

**7.Bounce Rates:** Average bounce rate of the pages visited by the user. It represents the percentage of visitors who enter the site and then leave rather than continuing to view other pages within the same site.

**8.Exit Rates:** Average exit rate of the pages visited by the user. It is the percentage of people who left your site from that page.

**9.Page Values:** Average page value of the pages visited by the user. It is the average value for a page that a user visited before landing on the goal page or completing an Ecommerce transaction (or both).

**10.Special Day** Closeness of the visiting day to a special event like Motherâ€Ÿs Day or festivals like Christmas.

**11.Month:** Visiting month during the whole year.

**12.Operating Systems:** Operating Systems of the visitor.

**13.Browser:** Browser of the visitor.

**14.Region:** Geographic region from which the session has been started by the visitor.

**15.Traffic Type:** Traffic source through which user has entered the website.

**16.Visitor Type:** Defines the nature of user.

**17.Weekend:** Defines the timing of the user.

```python
# variance threshold 
from sklearn.feature_selection  import VarianceThreshold
```

```python
var_thres1 = VarianceThreshold(threshold=0)
var_thres1.fit(X_train)
```

```python
# Finding number of non constant features
sum(var_thres1.get_support())
```

```python
# looking for constant columns
constant_column1 = [column for column in X_train.columns
                    if column not in X_train.columns[var_thres1.get_support()]]
print(len(constant_column1))
```

We find one column with the constant value lets see which one itis and remove that

```python
for feature in constant_column1:
  print(feature)
```

We remove this column from our X table

```python
X.drop(constant_column1,axis=1)
```

After working on the constant using variance threshold  it is found there is no columns with constant 0 so we proceed further with the model building after applying variance threshold

##Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forest classifier

rfc_1 = RandomForestClassifier(n_estimators = 10,criterion="entropy",max_depth=None) 
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
rfc_1.fit(X_train, y_train)
 
# performing predictions on the test dataset
y_predt = rfc_1.predict(X_train)
y_pred1 = rfc_1.predict(X_test)

from sklearn.metrics import confusion_matrix
cm_train = confusion_matrix(y_train, y_predt)
cm = confusion_matrix(y_test, y_pred1)
print("Confusion Matrix of Train: ", cm_train)
print("Confusion Matrix of Test: ", cm)


 
# metrics are used to find accuracy or error
from sklearn import metrics 
print()
 
# using metrics module for accuracy calculation
print("ACCURACY OF THE Train MODEL: ", metrics.accuracy_score(y_train, y_predt))
print("ACCURACY OF THE Test MODEL: ", metrics.accuracy_score(y_test, y_pred1))


#Accuracy, Precision, Recall, and F1-score can altogether be calculated using the method classification_report in python
print("Metric classification report for train: ", metrics.classification_report(y_train, y_predt))

print("Metric classification report for test: ", metrics.classification_report(y_test, y_pred1))

```

We get 90% of accuracy without dropping any features using random forest classifier
